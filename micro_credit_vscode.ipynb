```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
# Standardization
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Normalization
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# Convert the scaled data back to a DataFrame
scaled_data = pd.DataFrame(scaled_data, columns=data.columns)
```

```python
num_data = scaled_data[["aon","daily_decr30","daily_decr90","rental30","rental90","last_rech_amt_ma","sumamnt_ma_rech30",
                 "medianmarechprebal30","sumamnt_ma_rech90"]]
```

```python
# Define outlier threshold multiplier (consider 1.5 for standard definition)
outlier_multiplier = 1.5

for column in num_data:
    # Calculate quartiles
    q1 = num_data[column].quantile(0.25)
    q3 = num_data[column].quantile(0.75)

    # Calculate interquartile range (IQR)
    iqr = q3 - q1

    # Define lower and upper bounds
    lower_bound = q1 - outlier_multiplier * iqr
    upper_bound = q3 + outlier_multiplier * iqr

    # Identify outliers
    outliers = num_data[(num_data[column] < lower_bound) | (num_data[column] > upper_bound)]


    # Option 2: Cap outliers to bounds
    num_data.loc[num_data[column] < lower_bound, column] = lower_bound
    num_data.loc[num_data[column] > upper_bound, column] = upper_bound

    # Option 3: Replace outliers with median/mean (consider carefully)
    scaled_data.loc[num_data[column] < lower_bound, column] = num_data[column].median()

    # Print information about outliers
    print(f"Column: {column}")
    print(f"Number of outliers detected: {len(outliers)}")
    print(f"IQR: {iqr}")
    print(f"Lower bound: {lower_bound}")
    print(f"Upper bound: {upper_bound}")
    print("================================================")
```

```python
# Calculate skewness for each column
from scipy.stats import skew
skewness = num_data.apply(lambda x: skew(x))

print("Skewness for each column:")
print(skewness)
```

## Random forest classifier

```python
x = scaled_data.iloc[:,1::]
y = scaled_data.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=17,test_size = 0.3)

from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
rf.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_rf))
```

```python
features = pd.DataFrame(rf.feature_importances_, index = x.columns)
print(features)
```
